## NFA to DFA Conversion

### Initialize Variables for the NFA

```python
alphabet = ("አን", "እ", "ነ", "ር", "ተ", "ቺ", "ሷ",
            "ሱ", "ሳቸው", "ስዎ", "ኔ", "ኛ", "ናንተ")
states = {"0", "1", "2", "3", "4", "5", "6"}
initial = "0"
final = "6"

nfa_transitions = {
    "0": {
        alphabet[0]: "1",
        alphabet[1]: {"2", "3"}
    },
    "1": {
        alphabet[4]: "6",
        alphabet[5]: "6"
    },
    "2": {
        alphabet[3]: "3",
        alphabet[2]: {"4", "5"},
        alphabet[10]: "6",
        alphabet[11]: "6",
        alphabet[12]: "6",
    },
    "3": {
        alphabet[6]: "6",
        alphabet[7]: "6",
        alphabet[8]: "6",
        alphabet[9]: "6",
    },
    "4": {
        alphabet[3]: "5"
    },
    "5": {
        alphabet[7]: "6"
    }
}
```

### Generate Diagram using `automata-lib`

```python
from automata.fa.nfa import NFA

# Initialize the NFA
nfa = NFA(
    states=states,
    input_symbols=alphabet,
    transitions=nfa_transitions,
    initial_state=initial,
    final_states={final},
)

# Display the NFA diagram
nfa.show_diagram(
    layout_method="dot",
    state_separation=0.65,
    arrow_size=0.85,
    font_size=14
)
```

### Convert from NFA to DFA with `automata-lib`

```python
from automata.fa.dfa import DFA

# Convert the NFA to a DFA
dfa = DFA.from_nfa(nfa)

# Display the DFA diagram
dfa.show_diagram(
    layout_method="dot",
    state_separation=0.65,
    arrow_size=0.85,
    font_size=14
)
```
